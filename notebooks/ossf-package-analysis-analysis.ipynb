```python
import json
import os
import pandas as pd
```

```python
random_results_dir = os.path.abspath("../results/ossf-package-analysis/random/")
```

```python
def analyze_results(results_dir):
    results = []
    for root, dirs, files in os.walk(results_dir):
        for file in files:
            file_path = os.path.join(root, file)
            if file_path.endswith(".json"):
                package_name = file_path.split("/")[7]                
                with open(file_path, "rb") as f:
                    data_bytes = f.read()
                    data_str = data_bytes.decode("utf-8", errors='ignore')
                    try:
                        data = json.loads(data_str)
                    except json.JSONDecodeError:
                        import pdb; pdb.set_trace()
                    analysis_results = data["Analysis"]["install"]
                    status = analysis_results["Status"]
                    if status == "completed":
                        import pdb; pdb.set_trace()
                        files_events = analysis_results["Files"]
                        commands_events = analysis_results["Commands"]
                        dns_events = analysis_results["DNS"]
                        results.append([package_name, len(files_events), len(commands_events), len(dns_events)])

    return results
```

```python
random_results = analyze_results(random_results_dir)
```

```python
random_results_df = pd.DataFrame(random_results, columns=["package", "num_files", "num_commands", "num_dns"])
random_results_df.head()
```

```python
print(f"Total number of files events: {random_results_df['num_files'].sum()}")
random_results_df["num_files"].describe()
```

```python
print(f"Total number of commands events: {random_results_df['num_commands'].sum()}")
random_results_df["num_commands"].describe()
```

```python
print(f"Total number of dns events: {random_results_df['num_dns'].sum()}")
random_results_df["num_dns"].describe()
```
